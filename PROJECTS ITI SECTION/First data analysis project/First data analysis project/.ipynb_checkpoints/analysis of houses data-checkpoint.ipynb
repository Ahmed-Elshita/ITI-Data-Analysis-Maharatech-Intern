Separate features from target field

```python
target = df['SalePrice']
df=df.drop(columns=['SalePrice'])
```

Separate numeric features from categorical features
then deal with missing values 
put columns mean value for numeric fields
put 'Missing' label to categorical fields

```python
df_numeric= df.select_dtypes(include=[np.number])
print(df_numeric.shape)
df_numeric.head()
```

```python
df_non_numeric= df.select_dtypes(exclude=[np.number])
print(df_non_numeric.shape)
df_non_numeric.head()
```

```python
df_numeric.isna().sum()
```

Fill null value (imputation)

```python
for col in df_numeric.columns:
    df[col].fillna(df[col].mean(),inplace=True)
    
for col in df_non_numeric.columns:
    df[col].fillna('Missing',inplace=True)    
    
df.isnull().sum()
```

# do scaling for numeric values

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_numeric_scale=sc.fit_transform(df_numeric)
df_numeric_scale = pd.DataFrame(x_numeric_scale,columns=df_numeric.columns)
df_numeric_scale.head()
```

# Get Outliers

```python
df_outlier=pd.DataFrame()
for col in df_numeric_scale.columns:
    df_outlier['Outlier_'+col]=df_numeric_scale[col].apply(lambda x: 1 if  np.abs(x)>3 else 0)
df_outlier.sum()          
```

by boxplot

```python
df_numeric.boxplot(column='LotFrontage')
plt.title('LotFrontage')
plt.ylabel('LotFrontage')
plt.show()
```

```python
for col in df_numeric.columns:
    Q1=df_numeric[col].quantile(.25)
    Q3=df_numeric[col].quantile(.75)
    IQR = Q3 - Q1
    minimum= Q1 - 1.5 * IQR
    maximum= Q3 + 1.5 * IQR
    df_numeric[col]=df_numeric[col].apply(lambda x: replace_outlier(x,minimum,maximum) )
df_numeric.head(10)
```
